iki basamaklı ikiz asal sayıların (aralarındaki fark 2 olan asal sayıların ) en büyük çiftinibulan fonksiyonu hazırlayınız. 
Örn:

*    en küçük ikiz asal sayılar 3 ve 5
*   en küçük iki basamaklı asal sayılar 11 ve 13

```python
def asalSayiKontrol(sayi):
  kontrol=0
  for i in range(2,sayi):
    if sayi%i==0:
      kontrol=1
      break
  if kontrol==0:
    # print(sayi,'asaldır')
    return sayi
  else:
  #   print(sayi,'asal degildir')
    return 0
```

Harf Notu, Katsayı

```python
sayacA=0
sayacB=0
sayacC=0
sayacD=0
sayacF=0
hesapla = 0
harf = input('Harf notunu giriniz(A,B,C,D veya F): ')
while harf != 'H' and harf != 'h':
  if harf== 'A' or harf== 'a':
    sayacA +=1
    hesapla += 4
  elif harf == 'B' or harf== 'b':
    sayacB += 1
    hesapla += 3
  elif harf == 'C' or harf== 'c':
    sayacC += 1
    hesapla += 2
  elif harf == 'D' or harf== 'd':
    sayacD += 1
    hesapla += 0.5
  elif harf == 'F' or harf== 'f':
    sayacF += 1    
  else:
    print('Hatalı kod girdiniz yeniden deneyiniz!!')
  harf = input('Harf notunu giriniz(A,B,C,D veya F): ')

ort = hesapla / (sayacA+sayacB+sayacC+sayacD+sayacF)
print('Kredi ortalamanız: ',ort)
print('Toplam girilen ders sayısı: ', (sayacA+sayacB+sayacC+sayacD+sayacF))
print('A: ',sayacA,' B: ',sayacB,' C: ',sayacC,' D: ',sayacD,' F: ',sayacF)
```

```python
def tansiyonKontrol(tansiyon,t_altsinir,t_ustsinir,dusuk,normal,yuksek):  
  if tansiyon< t_altsinir:    
    dusuk +=1
  elif tansiyon <= t_ustsinir and tansiyon >= t_altsinir :
    normal +=1
  else:   
    yuksek +=1
  return dusuk,normal,yuksek
```

Bir odada bulunan kişi sayısına göre odada bulunan iki kişişnin doğum günlerinin aynı olma ihtimali aşağıdaki gibi hesaplanır.

N=odadaki kişi sayısı olmak üzere
Olasılık=1 - (365/366). (364/366). (363/366) ...(366-N+1/366)
Klavyeden girilen kişi sayısına göre olasılığı hesaplayan fonksiyonu gerçekleştiriniz.

```python
def dogumG(n):  
  olasilik1=1
  for i in range(1,n+2):   
    olasilik1*=((366-i)/366)
  return (1-olasilik1)*100

n=int(input('Odadaki kişi sayısını giriniz: '))
print(dogumG(n))
```
