Ekrandan istenen kalma süresine göre otopark ücretini hesaplayıp ekrana yazdıran kodu tasarlayınız.

+ Süre < 1 saat ise ücret:5TL
+ 1<= süre < 5 saat ise ücret:10TL
+ 5<= süre < 10 saat ise ücret:17TL
+ Süre >=10 ise 10. saatten sonraki her saat için 5TL

```python
sure = float(input('Otoparkta kalma süresini giriniz: '))
if sure < 1:
  print('Ucret 5 TL')
elif sure >=1 and sure <5:
  print('Ucret 10 TL')
elif sure >=5 and sure <10:
  print('Ucret 17 TL')
elif sure >=10:
  hesap = ((int(sure) - 10)*5) + 17
  print("Ucret " + str(hesap) + " TL")
else:
  print('Geçerli bir süre giriniz!!')
```
